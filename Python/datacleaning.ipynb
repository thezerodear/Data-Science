```python
def value_unit(s: str):
    s = s.strip()
    if "พัน" in s:
        return 1000
    else:
        return 1
```

```python
value_unit("10พัน")
```

```python
import re
def str_value(s: str):
    if isinstance(s, str):
        mul = value_unit(s)
        return float(re.sub(r'[^0-9.]', '', s)) * mul
    else:
        return s
```

```python
str_value("12.5พัน")
```

```python
def month(s: str):
    try:
        if "ปี" in s:
            numeric_value = float(s.replace("ปี", "").strip())
            return numeric_value * 12
        elif "วัน" in s:
            numeric_value = float(s.replace("วัน", "").strip())
            return numeric_value / 30
        else:
            numeric_value = float(s.replace("เดือน", "").strip())
            return numeric_value
    except ValueError:
        return "ไม่สามารถแปลงเป็นตัวเลขได้"

```

```python
month("12วัน")
```

```python
df['join']=df['join'].apply(lambda x:month(x))
```
