```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.imdb.com/chart/top/'


headers={"User-Agent": "Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148"}

resp = requests.get(url, headers=headers)
imdb = BeautifulSoup(resp.text,'html.parser')
```

```python
rawmovie = imdb.find_all('div',{'class':'ipc-metadata-list-summary-item__tc'})
```

```python
mname = rawmovie[0].find('h3').get_text()
```

```python
meta = {'year':None,'Length':None,'Rating':None}
```

```python
mrate = rawmovie[0].find_all('span',{'class':'sc-1e00898e-8 hsHAHC cli-title-metadata-item'})
if len(mrate) >= 3:
    meta['year'] = mrate[0].get_text()   # ปี
    meta['Length'] = mrate[1].get_text() # ระยะเวลา
    meta['Rating'] = mrate[2].get_text()
    meta['titles']= mname
```

```python
results = []  # สำหรับเก็บข้อมูลทั้งหมด

for movie in rawmovie:
    mrate = movie.find_all('span', {'class': 'sc-1e00898e-8 hsHAHC cli-title-metadata-item'})
    
    if len(mrate) >= 3:
        meta = {}
        meta['year'] = mrate[0].get_text()   # ปี
        meta['Length'] = mrate[1].get_text() # ระยะเวลา
        meta['Rating'] = mrate[2].get_text() # เรตติ้ง

        
        meta['titles'] = movie.find('h3').get_text()

        results.append(meta)

# ตรวจสอบผลลัพธ์
for result in results:
    print(result)
```

```python
import pandas as pd
imdb = pd.DataFrame(results)
```

```python
import re
```

```python
imdb['titles'] = imdb['titles'].str.replace(r'^\d+', "", regex=True)
```
